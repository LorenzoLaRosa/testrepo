```python
# Standard Python libraries
import os
import datetime
import sys
import re

# https://pandas.pydata.org/docs/
import pandas as pd

# http://www.numpy.org/
import numpy as np
np.set_printoptions(precision=4, suppress=True)

# https://github.com/usnistgov/atomman
import atomman as am
import atomman.unitconvert as uc

# https://matplotlib.org/
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['text.usetex'] = True

# Show atomman version
print('atomman version =', am.__version__)

# Show date of Notebook execution
print('Notebook executed on', datetime.date.today())
```

```python
# Read the EAM02_thermo_output.txt

thermo_output = "EAM02_thermo_output.txt"

with open(thermo_output, "r") as file1:
    FileasList = file1.readlines()
    
data = FileasList[2::3]

# Write the lammps_output.txt
#with open('2NNMEAM_lammps_output.txt','w') as writefile:
#    for line in data:
#        writefile.write(line)
        
# Define lammps dictionary 'lammps_dict'

lammps_dict = {'x-iter': [], 'y-iter': [], 'E_i [eV]': [], 'E_f [eV]': [], 'area': [], 'SFE [mJ/m^2]': []}

for i in range(0, len(data)):
    split_array = re.split("\s", data[i]) # Use the split function to split the string by the "\s"
    split_array.remove(split_array[-1])

    lammps_dict['x-iter'].append(float(split_array[0])/20) # Append to the key x-iter
    lammps_dict['y-iter'].append(float(split_array[1])/20) # Append to the key y-iter
    lammps_dict['E_i [eV]'].append(float(split_array[2])) # Append to the E_i [eV]
    lammps_dict['E_f [eV]'].append(float(split_array[3])) # Append to the E_f [eV]
    lammps_dict['area'].append(float(split_array[4])) # Append to the key area
    lammps_dict['SFE [mJ/m^2]'].append(float(split_array[5])) # Append to the key SFE [mJ/m^2]
    
del(i) # Delete loop variable

# Casting the dictionary to a DataFrame
df = pd.DataFrame(lammps_dict)

# Reconstruct dataframe to have fully periodic solutions
df_block = df.iloc[-20:, :]
new_row = df.iloc[[-1]]

df_EAM02_MD = pd.concat([new_row, df_block]).reset_index(drop = True)
df_EAM02_MD['y-iter'][0]=0

for i in range(0, len(df_EAM02_MD['x-iter'])):
    df_EAM02_MD['x-iter'][i] = 0
del(i) # Delete loop variable
    
for i in range(0,20):
    df_block = df.iloc[i*20:(i*20)+20, :]
    new_row = df_block.iloc[[-1]]
    
    df_block = pd.concat([new_row, df_block]).reset_index(drop = True)
    df_block['y-iter'][0]=0
    
    df_EAM02_MD = pd.concat([df_EAM02_MD, df_block]).reset_index(drop = True)
del(i) # Delete loop variable

# Save to Excel file
df_EAM02_MD.to_excel("EAM02_LAMMPS_output.xlsx")
```
